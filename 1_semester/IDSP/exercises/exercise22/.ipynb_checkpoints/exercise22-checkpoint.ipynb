## 22.2

Implement your own function to compute the Degree centrality and test it on the karate club graph

```python
# Graph used to test functions
G = nx.karate_club_graph()

#print(G.nodes)

def degree(graph):
    degrees = {}
    nodes = graph.nodes
    for i in nodes:
        neighbors = list(graph.neighbors(i))
        degree = len(neighbors)
        degrees[i] = degree / (len(nodes) - 1)
    return degrees

degree(G)
```
